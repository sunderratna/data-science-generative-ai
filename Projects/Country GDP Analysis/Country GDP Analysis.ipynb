```python
# Checking insights, patterns, and hidden details through advanced statistics and visualization

import matplotlib.pyplot as plt  # visualization
import seaborn as sns  # statistics visualization (advanced visualization)

# To set the plot inside the notebook
%matplotlib inline
plt.rcParams['figure.figsize'] = 6, 3

import warnings
warnings.filterwarnings('ignore')
```

```python
# Display the first few rows again
df.head()
```

```python
# Distribution plot (univariate analysis)
vis1 = sns.distplot(df['InternetUsers'])
```

```python
# Distribution plot with specified number of bins
vis2 = sns.distplot(df['InternetUsers'], bins=10)
```

```python
# Displot (univariate analysis without distribution line)
vis3 = sns.displot(df['InternetUsers'])
```

```python
# Displot with bins (without distribution line)
vis4 = sns.displot(df['InternetUsers'], bins=10)
```

```python
# Box plot (bivariate analysis between BirthRate and InternetUsers)
vis5 = sns.boxplot(data=df, x='InternetUsers', y='BirthRate')
```

```python
# Linear model plot with regression line
vis6 = sns.lmplot(data=df, x='InternetUsers', y='BirthRate')
```

```python
# Linear model plot without regression line
vis7 = sns.lmplot(data=df, x='InternetUsers', y='BirthRate', fit_reg=False)
```

```python
# Linear model plot with hue (color-coded by 'IncomeGroup') and without regression line
vis8 = sns.lmplot(data=df, x='InternetUsers', y='BirthRate', hue='IncomeGroup', fit_reg=False)
```

```python

# Linear model plot with hue and regression line
vis9 = sns.lmplot(data=df, x='InternetUsers', y='BirthRate', fit_reg=True, hue='IncomeGroup')
```
