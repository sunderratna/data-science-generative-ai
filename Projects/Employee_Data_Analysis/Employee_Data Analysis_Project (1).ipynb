```python
# Importing additional libraries for data visualization
import numpy as np  # Library for numerical computations
import matplotlib.pyplot as plt  # Library for plotting
import seaborn as sns  # Library for statistical data visualization
```

```python
# Creating a distribution plot of the SALARY column
vis1 = sns.histplot(emp['SALARY'])
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Creating a distribution plot of the SALARY column using a different method
vis1 = sns.displot(emp['SALARY'])
plt.title('Distribution of Salaries')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.show()
```

```python
# Creating a scatter plot with a regression line of Experience vs Salary
vis2 = sns.lmplot(data=emp, x='EXP', y='SALARY')
plt.title('Experience vs Salary with Regression Line')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.show()
```

```python
# Creating a scatter plot of Experience vs Salary without a regression line
vis2 = sns.lmplot(data=emp, x='EXP', y='SALARY', fit_reg=False)
plt.title('Experience vs Salary')
plt.xlabel('Experience (Years)')
plt.ylabel('Salary')
plt.show()
```
